```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("E:\Python project\Zomato Data Analysis Using Python\Zomato data .csv")
print(df.head())
```

```python
def handleRate(value):
    value=str(value).split('/')
    value=value[0];
    return float(value)
```

```python
df['rate']=df['rate'].apply(handleRate)
print(df.head())
```

```python
df.info()
```

```python
custom_palette = ['#FF6347', '#4682B4', '#32CD32', '#FFD700', '#6A5ACD', '#FF69B4', '#8A2BE2', '#00CED1', '#FF4500', '#2E8B57']
plt.figure(figsize=(12, 8))
sns.countplot(x='listed_in(type)', data=df, palette=custom_palette[:len(df['listed_in(type)'].unique())])
plt.xlabel("Type of Restaurant")
plt.ylabel("Count")
plt.title("Count Plot of Restaurant Types")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
grouped_data = df.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c="green", marker="o")
plt.xlabel("Type of restaurant", c="red", size=20)
plt.ylabel("Votes", c="red", size=20)
```

```python
max_votes = df['votes'].max()
restaurant_with_max_votes = df.loc[df['votes'] == max_votes, 'name']
 
print("Restaurant(s) with the maximum votes:")
print(restaurant_with_max_votes)
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='online_order', data=df, palette='Set2')
plt.xlabel('Online Order')
plt.ylabel('Count')
plt.title('Count Plot of Online Orders')
plt.show()
```

```python
plt.hist(df['rate'],bins=5)
plt.title("Ratings Distribution")
plt.show()
```

```python
couple_data=df['approx_cost(for two people)']
custom_palette = ['#FF6347', '#4682B4', '#32CD32', '#FFD700', '#6A5ACD', '#FF69B4']
sns.countplot(x=couple_data, palette=custom_palette)
```

```python
custom_palette = ['#FF6347', '#4682B4']
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rate', data = df, palette=custom_palette)
```

```python
pivot_table = df.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```
